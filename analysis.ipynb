# Bivariate Analysis #

#### Before diving into bivariate analysis, it's crucial to have a preliminary understanding of how different variables relate to each other. In my case, I identified that soil water content and soil pH may be influenced by factors like slope and elevation. This observation helps me narrow down the focus of the bivariate analysis. ####

#### By doing this initial exploration, I've reduced the scope of my analysis to the variables that are most likely to exhibit meaningful relationships with each other. This targeted approach can save time and resources compared to performing bivariate analysis on the entire dataset. ####

```python
import matplotlib.pyplot as plt

# Define the variables
bangor_distance = df_bangor['Distance..m.']
bangor_pH = df_bangor['pH']
bangor_water_content = df_bangor['Water.Content']

peris_distance = df_peris['Distance..m.']
peris_pH = df_peris['pH']
peris_water_content = df_peris['Water.Content']

# Create the figure and the first y-axis (for pH)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Set labels and titles
axes[0].set_xlabel('Distance')
axes[0].set_ylabel('pH', color='tab:blue')
axes[0].set_title('Bangor - pH and Water Content')

axes[1].set_xlabel('Distance')
axes[1].set_ylabel('pH', color='tab:blue')
axes[1].set_title('Peris - pH and Water Content')

# Plot pH on the left axis
axes[0].plot(bangor_distance, bangor_pH, color='tab:blue', marker='o', label='pH')
axes[1].plot(peris_distance, peris_pH, color='tab:blue', marker='o', label='pH')

# Create a twin y-axis for water content on the right side
ax1_water_content = axes[0].twinx()
ax2_water_content = axes[1].twinx()

# Plot water content on the right axis
ax1_water_content.plot(bangor_distance, bangor_water_content, color='tab:green', marker='x', label='Water Content')
ax2_water_content.plot(peris_distance, peris_water_content, color='tab:green', marker='x', label='Water Content')

# Set y-axis labels and colors for water content
ax1_water_content.set_ylabel('Water Content', color='tab:green')
ax2_water_content.set_ylabel('Water Content', color='tab:green')

# Adjust layout
fig.tight_layout()

# Show the plot
plt.show()
```
